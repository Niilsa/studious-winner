```python
image_path = '/home/niilsa/Pictures/lucy.jpg'
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Read the image file
image = plt.imread(image_path)
```

```python
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        for k in range(image.shape[2]):
            if image[i][j][k] < 30:
                image[i][j][k] = 0
```

```python
plt.imshow(image)
plt.show()
```
